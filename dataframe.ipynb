## create an Data Frame in Pandas

```python
import pandas as pd
```

```python
data={
     "rollno":[111,222,333],
    "students": ["ram","shyam", "Rameswaram"],
    "marks":[24,45,67],
    "age":[12,12,34]
}
```

```python
data2={
    "rollno.":[111,222,333,555],
    "subject":["maths","bio","Hindi","GioGraphy"]
    
}
```

```python
d= pd.DataFrame(data) 

print("datasets",d)
```

```python
d2= pd.DataFrame(data2) 

print("datasets \n",d2)
```

### access the dataframe data by loc()

```python
print(d.loc[1,"age"])
```

```python
print(d.iloc[[0,1,2]])
```

### how to set indexes by myself

```python
df= pd.DataFrame(data,index=["s1","s2","s3"])

print(df)
```

Print the column name through iteration

```python
for row in df:
    print(row)
```

### Pandas Series

```python
data3=[1,2,3,4,5,32]
```
